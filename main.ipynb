```python
import cv2
import matplotlib.pyplot as plt
import numpy as np 
```

```python
img = cv2.imread('orginal_image.webp')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
plt.title('Orginal image ')
plt.show()
```

```python
blur_img = cv2.blur(img, (5,5))
plt.imshow(blur_img)
plt.title('blured Image ')
plt.show()
```

```python
median = np.median(blur_img)
lower_threshold = int(max(0, 0.5 * median))
upper_threshold = int(min(255, 1/3 * median))
```

```python
canny_edg = cv2.Canny(blur_img, lower_threshold, upper_threshold)
plt.imshow(canny_edg)
plt.title("Edges detected using canny_edge")
plt.show()
```
